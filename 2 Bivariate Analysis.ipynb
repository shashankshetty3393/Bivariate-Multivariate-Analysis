# doing bivariate analysis
> find what type of data it is
> there are three chances
1) numerical numerical data analysis(scatterplot)
2) numerical categorical analysis(boxplot,distplot)
3) categorical categorical analysis

# 1. Scatterplot (numerical numerical) 

```python
#bivariate on tips data
sns.scatterplot(x=tips['total_bill'],y=tips['tip'])
```

```python
#multi variate
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],style=tips['sex'],hue=tips['smoker'],size=tips['size'])
```

# Barplot (num cat datatype)

X axis = categorical
Y axis = Numerical

```python
# on titanic dataset
titanic.head()
```

```python
#bivariate
sns.barplot(x=titanic['Pclass'],y=titanic['Fare'])
```

```python
sns.barplot(x=titanic['Pclass'],y=titanic['Age'])
```

```python
#multivariate
sns.barplot(x=titanic['Pclass'],y=titanic['Fare'],hue=titanic['Sex'])
```

```python
sns.barplot(x=titanic['Pclass'],y=titanic['Age'],hue=titanic['Sex'])
```

# Boxplot (Num cat) mathematical comparision

```python
sns.boxplot(titanic['Age'])
```

```python
sns.boxplot(x=titanic['Sex'],y=titanic['Age'])
```

```python
sns.boxplot(x=titanic['Sex'],y=titanic['Age'],hue=titanic['Survived'])
```

# Distplot ( Num Cat)  finding probability distribution function

```python
#finding pdf on survied titanic dataset 0=dead
sns.distplot(titanic[titanic['Survived']==0]['Age'])
```

```python
sns.distplot(titanic[titanic['Survived']==1]['Age'])
```

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'])
sns.distplot(titanic[titanic['Survived']==1]['Age'])
```

# Heatmap ( Cat cat)

```python
titanic.head()
```

```python
#finding how many survived according to pclass
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
titanic['Survived'] = pd.to_numeric(titanic['Survived'], errors='coerce')
titanic.groupby('Pclass')['Survived'].mean()*100
```

```python
(titanic.groupby('Pclass')['Survived'].mean()).plot(kind='bar')
```

```python
titanic.groupby('Embarked')['Survived'].mean()*100
```

# Cluster map (cat cat: relative group plot together)

```python
pd.crosstab(titanic['Parch'],titanic['Survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```

# Lineplot : joining all the scatterplot point is called lineplot

> we need to use lineplot when we have x axis quantity is time based(year,date,month)
> *stocks
>  

```python
flights.head()
```

```python
flights.groupby('year').sum(numeric_only = True)
```

```python
new = flights.groupby('year').sum(numeric_only = True).reset_index()
new
```

```python
sns.lineplot(x=new['year'],y=new['passengers'])
```

```python
flights.pivot_table(values='passengers',index='month',columns='year')
```

```python
sns.heatmap(flights.pivot_table(values='passengers',index='month',columns='year'))
```

```python
sns.clustermap(flights.pivot_table(values='passengers',index='month',columns='year'))
```
